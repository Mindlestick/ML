```python
import pandas as pd

train=pd.read_csv("/content/train_set.csv")
y=pd.DataFrame(train["LABEL"])
del train["LABEL"]
x=train
x_test=pd.read_csv("/content/x_test.csv")
del x["TIMESTAMP"]
del x_test["TIMESTAMP"]
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in y:
  y[i]=encoder.fit_transform(y[i])
```
