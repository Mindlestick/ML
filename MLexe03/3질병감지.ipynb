```python
def Df_Trans(df):
  from sklearn.preprocessing import LabelEncoder
  le = LabelEncoder()
  for key in df.keys():
    df[key] = le.fit_transform(df[key])

  #print(le.classes_)
  return df
```

```python
import pandas as pd

data = pd.read_csv("/content/train_set.csv")
y = data["lumpy"]
del data["lumpy"]
x = data
#x=x.dropna()
#del x["reportingDate"]
#del x["country"]
#del x["region"]
x=Df_Trans(x)
x
```

```python
test = pd.read_csv("/content/test_set.csv")
#test=test.dropna()
#del test["reportingDate"]
#del test["country"]
#del test["region"]
test=Df_Trans(test)
test
```

```python
import xgboost
model = xgboost.XGBClassifier()
model.fit(x,y)
pred = model.predict(test)
print(pred)

sub = pd.read_csv("/content/sample_submission.csv")
sub["lumpy"] = pred

sub
sub.to_csv("sub.csv",index=False)
```
