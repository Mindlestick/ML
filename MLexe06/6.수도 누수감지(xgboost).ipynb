```python
import numpy as np
import pandas as pd

x_test=pd.read_csv("/content/x_test.csv")
train=pd.read_csv("/content/train_set.csv")
sub = pd.read_csv('/content/sample_submission.csv')
```

```python
y=train["leaktype"]
y = pd.DataFrame(y)
y
```

```python
del train["leaktype"]
x=train
x
```

```python
train.shape
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for i in y:
   y[i] = encoder.fit_transform(y[i])
```

```python
del x["ldate"]
del x_test["ldate"]
```

```python
y
```

```python
import xgboost

model = xgboost.XGBClassifier(max_depth=10,n_estimators=100)
model.fit(x,y)
pred = model.predict(x_test)
pred
```

```python
inverse_pred = encoder.inverse_transform(pred)
print(inverse_pred)
```

```python
sub = pd.read_csv('/content/sample_submission.csv')
sub["leaktype"] = inverse_pred
```

```python
sub.to_csv("sub.csv",index=False)
```
