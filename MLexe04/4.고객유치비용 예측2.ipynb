```python
import pandas as pd

x=pd.read_csv("/content/x_train.csv")
y=pd.read_csv("/content/y_train.csv")
x_test=pd.read_csv("/content/x_test.csv")
sub=pd.read_csv("/content/sample_submission.csv")
x
```

```python
def Label(df):
  from sklearn.preprocessing import LabelEncoder
  encoder = LabelEncoder()
  for i in df:
    df[i] = encoder.fit_transform(df[i])
```

```python
Label(x)
Label(x_test)
x
```

```python
# 정확도 별로 안높음

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(x,y)
pred = model.predict(x_test)

#print(model.score(x,y))
pred
```

```python
# 0.9472
import xgboost

model = xgboost.XGBRegressor(max_depth=13,n_estimators=90,learning_rate=0.5)
model.fit(x,y)
pred = model.predict(x_test)

pred
```

```python
# 측정 안해봄
import xgboost

model = xgboost.XGBRegressor(max_depth=13,n_estimators=240,learning_rate=0.45)
model.fit(x,y)
pred = model.predict(x_test)

pred
```
