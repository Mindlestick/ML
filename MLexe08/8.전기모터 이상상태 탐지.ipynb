```python
import pandas as pd
import numpy as np
x0=pd.read_csv("/content/x_train_abnormal.csv")
x1=pd.read_csv("/content/x_train_normal.csv")
x_test=pd.read_csv("/content/x_test.csv")
sub=pd.read_csv("/content/sample_submission.csv")
```

```python
x_test
```

```python
def T(df):
  new_df = pd.DataFrame()

  for _ in range(1000):
    new_df[str(_)] = []

  for i in range(1104):
    a=list()
    for j in range(1000):
      a.append(df['0'][i*1000+j])
    new_df.loc[i] = a

  return new_df
```

```python
def Statistics(new_df):
  new_dff=pd.DataFrame()

  new_dff['mean'] = new_df.mean(axis=1)
  new_dff['max'] = new_df.max(axis=1)
  new_dff['min'] = new_df.min(axis=1)
  new_dff['median'] = new_df.median(axis=1)
  #new_dff['mode'] = new_df.mode(axis=1)
  new_dff['variance'] = new_df.var(axis=1)
  new_dff['std'] = new_df.std(axis=1)
  #new_dff['rms'] = new_df.rms(axis=1)
  new_dff['skew'] = new_df.skew(axis=1)
  new_dff['kurtosis'] = new_df.kurtosis(axis=1)
  new_dff['P_P'] = abs(new_df.max(axis=1)-new_df.min(axis=1))

  return new_dff
```

```python
new_x0=T(x0)
new_x1=T(x1)
```

```python
new_x00 = Statistics(new_x0)
new_x11 = Statistics(new_x1)
new_test = Statistics(x_test)
```

```python
result= pd.DataFrame()
result = pd.concat([new_x00,new_x11])
result

label = [0] * 1104 + [1] * 1104
result
```
